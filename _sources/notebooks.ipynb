```python
import pandas as pd
import math
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/FajarFatha/datamining/main/ds_salaries.csv")
```

```python
df.head()
```

## Proximity Measure for Nominal Attributes

```python
def d1(i,j):
    P = 8
    M = 0
    for col in ["work_year", "experience_level", "employment_type","job_title","salary_currency","employee_residence","company_location","company_size"]:
        if df[col][i] == df[col][j]:
            M += 1

    return (P-M)/P
```

```python
print(d1(1,2))
print(d1(1,3))
print(d1(1,4))
```

##Distance on Numeric Data: Minkowski Distance

```python
def d2(i,j, h):
    result = 0
    for col in ["salary", "salary_in_usd","remote_ratio"]:
        result = math.pow(abs(df[col][i] - df[col][j]), h)
    return result**(1/h)
```

Manhattan distance

```python
print(d2(1,2,1))
print(d2(1,3,1))
print(d2(1,4,1))
```

Euclidean distance

```python
print(d2(1,2,2))
print(d2(1,3,2))
print(d2(1,4,2))
```
